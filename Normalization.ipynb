# Normalization brain and ICV masks from Pincram

To normalize the "normal" brain masks, just change the variable icv_masks and output_dir

```python
import glob
```

```python
icv_masks = glob.glob('./pincram_all/icv/*.nii.gz')
```

```python
icv_masks
```

```python
output_dir = './Pincram_all_icv_normalized'
ref_file = '/usr/share/data/fsl-mni152-templates/MNI152_T1_1mm_brain.nii.gz'
```

```python
from nipype.interfaces import fsl
import os
```

```python
for icv in icv_masks:
    sub_id = os.path.basename(icv).split('.')[0]
    print('--- {0} ---'.format(sub_id))
    flt = fsl.FLIRT(bins=640, cost_func='mutualinfo')
    flt.inputs.in_file = icv
    flt.inputs.reference = ref_file
    flt.inputs.out_file = os.path.join(output_dir, sub_id + '.nii.gz')
    flt.inputs.out_matrix_file = os.path.join(output_dir, sub_id + '.mat')
    res = flt.run()
```
