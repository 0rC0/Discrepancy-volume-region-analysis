```python
base_path = '~/Documents/Praktikum/pincram_all'
icv_path = '~/Documents/Praktikum/pincram_all/icv'
brain_path = '~/Documents/Praktikum/pincram_all/pincram'
dr_path = '~/Documents/Praktikum/dr_labels'
```

```python
file_names = [f for f in os.listdir(brain_path) if f.endswith('nii.gz')]
file_names
```

```python
def set_zero_voxel(img ,dest, coord):
    # set to 0 voxels with coordinates from a tuple and save the img
    data = img.get_data()
    for t in coord:
        x,y,z = t
        data[x,y,z] = 0
    nib.save(img, dest)
    
def non_zero_voxels(img):
    # return a tuple of coordinates von voxel != 0
    x_dim, y_dim, z_dim = img.shape[0:3]
    data = img.get_data()
    l = []
    for x in range(x_dim):
        for y in range(y_dim):
            for z in range(z_dim):
                if data[x,y,z] != 0:
                    l.append((x,y,z))
    return tuple(l)
```

```python
for f in file_names:
    print(f)
    brain = os.path.join(brain_path, f)
    icv = os.path.join(icv_path, f)
    brain_mask = nib.load(brain)
    icv_file = nib.load(icv)
    dr_file = os.path.join(dr_path, f)
    #if the shapes are not the same there is a problem
    if icv_file.shape == brain_mask.shape:
        print(brain_mask.shape[0:3])
        t = non_zero_voxels(brain_mask)
        set_zero_voxel(icv_file, dr_file, t)
    else:
        print('Error with {0}'.format(f))
```
